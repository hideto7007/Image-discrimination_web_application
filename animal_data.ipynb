# データの前処理

```python
from PIL import Image
import os, glob
import numpy as np
from sklearn.model_selection import train_test_split
import pandas as pd

pd.options.display.max_columns = None
#ImageFile.LOAD_TRUNCATED_IMAGES = True

classes = ["monkey", "boar", "crow"]
num_classes = len(classes)
image_size = 50
num_testdata = 100

# 画像の読み込み

X_train = []
X_test  = []
Y_train = []
Y_test  = []


for index, classlabel in enumerate(classes): # classesリストから取得する
    photo_dir = "./" + classlabel           # フォルダー指定
    files = glob.glob(photo_dir + "/*.jpg") # フォルダー内の画像指定
    for i, file in enumerate(files):
        if i > 200:                         # 250枚までで学習を終了させる
            break
        image = Image.open(file)
        image = image.convert("RGB")
        image = image.resize((image_size, image_size))
        data = np.asarray(image)
       
        if i < num_testdata:
            X_test.append(data)
            Y_test.append(index)
        else:
            #X_train.append(data)
            #Y_train.append(index)
            
            for angle in range(-20, 20, 5):
                img_r = image.rotate(angle)
                data = np.asarray(img_r)
                X_train.append(data)
                Y_train.append(index)
                
                img_trans = image.transpose(Image.FLIP_LEFT_RIGHT)
                data = np.asarray(img_trans)
                X_train.append(data)
                Y_train.append(index)
        
#X = np.array(X)
#Y = np.array(Y)

# ラベル 0～200が0、201～402が1、403～は2

X_train = np.array(X_train)
X_test = np.array(X_test)
Y_train = np.array(Y_train)
Y_test = np.array(Y_test)

#print(X_train.shape)
#print(Y_train.shape)
#print(X_test.shape)
#print(Y_test.shape)


#X_train, X_test, Y_train, Y_test = train_test_split(X, Y)  # npy.ファイル保存時には使用しない
XY = (X_train, X_test, Y_train, Y_test)
np.save("./npy_file/animal_aug_1.npy", XY)
```
